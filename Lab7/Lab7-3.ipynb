```
! wget -c 'https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz'
! gunzip GoogleNews-vectors-negative300.bin.gz

from gensim.models import KeyedVectors
model1 = KeyedVectors.load_word2vec_format(r'GoogleNews-vectors-negative300.bin', binary=True)

print('The 10 words most similar to "teacher" are:')
print(model1.most_similar('teacher'))
print()
print('The 10 words most similar to "one" are:')
print(model1.most_similar('one'))
print()
print('King - Man + Woman equals:')
print(model1.most_similar(positive=['king', 'woman'], negative=['man'], topn=1))
print()
print('Paris - France + Germany equals:')
print(model1.most_similar(positive=['paris', 'germany'], negative=['france'], topn=1))
print()
print('The word vector for "one":')
print(model1['one'])
print()
print('The word vector for "notaword":')
print(model1['notaword'])
```
