```
from tensorflow import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data(num_words=20000)
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=80)
x_test = keras.preprocessing.sequence.pad_sequences(x_test, maxlen=80)

model = keras.models.Sequential()
model.add(keras.layers.Embedding(20000, 32))
model.add(keras.layers.LSTM(32))
model.add(keras.layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.fit(x_train, y_train, validation_data=(x_test, y_test),
        batch_size=32,
        epochs=5)
_, accuracy = model.evaluate(x_test, y_test, batch_size=32)
print('Test accuracy:', accuracy)
```
