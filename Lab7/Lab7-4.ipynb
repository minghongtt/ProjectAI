```
! wget -c 'https://dl.fbaipublicfiles.com/fasttext/vectors-wiki/wiki.en.zip'
! unzip wiki.en.zip wiki.en.bin

from gensim.models import FastText
model2 = FastText.load_fasttext_format('wiki.en.bin')

print('The 10 words most similar to "teacher" are:')
print(model2.wv.most_similar('teacher'))
print()
print('The 10 words most similar to "one" are:')
print(model2.wv.most_similar('one'))
print()
print('King - Man + Woman equals:')
print(model2.wv.most_similar(positive=['king', 'woman'], negative=['man'], topn=1))
print()
print('Paris - France + Germany equals:')
print(model2.wv.most_similar(positive=['paris', 'germany'], negative=['france'], topn=1))
print()
print('The word vector for "one":')
print(model2.wv['one'])
print()
print('The word vector for "notaword":')
print(model2.wv['notaword'])
```
